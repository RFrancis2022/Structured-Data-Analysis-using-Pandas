## Pivot Table

```python
a_ex = air_quality.pivot_table(values="value", index="location",columns="parameter",aggfunc="mean")

## values= "value" -  The column whose values we want to aggregate
## index= "location" - This sets the rows of the pivot table to be different locations (eg.station names or cities)
## columns = "paramter" - This sets the pivot table to be different paramters (NO2, PM10, etc).
## aggfunc = "mean" This tells pandas to compute the mean of the values for each combination of location and parameter.
```

```python
a_ex
```

```python
air_quality.pivot_table(
    values = "value",
    columns = "parameter",
    index = "location",
    aggfunc = "mean",
    margins = True,
)
## What margins=True does here: It adds an extra row and column labeled "All" to the pivot table,
## shows the overall mean across all locations and parameters
```

#### Pivot table is indeed indirectly linked to groupby(). The same result can be derived by grouping on both paramter and location:

```python
air_quality.groupby(["parameter","location"]) [["value"]].mean()
```
